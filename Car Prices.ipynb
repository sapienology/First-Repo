```python
import pandas as pd
```

```python
df1 = pd.read_csv('/content/cars.csv')
```

THE MODEL COLUMN

```python
df1['Model'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#df1['Model'] = df1['Model'].astype(str).str.strip()
```

```python
inte = LabelEncoder()
```

```python
df1['Model'] = inte.fit_transform(df1['Model'])
```

```python
df1['Model']
```

```python
df1['Model'].astype(int)
```

THE OWNER TYPE COLUMN

```python
df1['Owner_Type'].unique()
```

```python
df1['Owner_Type'] = df1['Owner_Type'].replace({
'First' : 0 ,
 'Second' : 1 ,
  'Third' : 2
})
```

```python
df1['Owner_Type'].astype(int)
```

THE TRANSMISION COLUMN

```python
df1['Transmission'].unique()
```

```python
df1['Transmission'] = df1['Transmission'].replace({
    'Manual' : 0 ,
    'Automatic' : 1
})
```

```python
df1['Transmission'].astype(int)
```

THE FUEL TYPE COLUMN

```python
df1['Fuel_Type'].unique()
```

```python
df1['Fuel_Type'] = df1['Fuel_Type'].replace({
    'Petrol' : 0 ,
    'Diesel' : 1
})
```

```python
df1['Fuel_Type'].astype(int)
```

THE YEAR COLUMN


```python
df1['Year'].unique()
```

THE BRAND COLUMN

```python
df1['Brand'].unique()
```

```python
df1['Brand'] = df1['Brand'].replace({
    'Toyota' : 0 ,
    'Honda'  : 1 ,
    'Ford'   : 2 ,
    'Maruti' : 3 ,
    'Hyundai' : 4 ,
    'Tata'   :   5 ,
    'Mahindra' : 6 ,
    'Volkswagen' : 7,
    'Audi'       : 8 ,
    'BMW'    : 9 ,
    'Mercedes' : 10
})
```

```python
df1['Brand'].astype(int)
```

```python
df1.info()
```

```python
df1 = df1.drop('Car_ID' , axis =1 )
```

```python
df1
```

```python
df1.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df1.drop(columns = ['Price'])
```

```python
Y = df1['Price']
```

```python
X_train ,X_test,Y_train ,Y_test = train_test_split(X ,Y ,test_size = 0.2 , random_state = 42)
```

```python
X_train
```

```python
df1['Price'].value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(X_train,Y_train)
```

```python
pred = log.predict(X_test)
```

```python
pred
```

```python
Y_test.values
```

```python
new_df = {'pred' : pred , 'Y_test' :Y_test}
```

```python
from sklearn.metrics import classification_report
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

print("MAE:", mean_absolute_error(Y_test, pred))
print("MSE:", mean_squared_error(Y_test, pred))
print("RMSE:", np.sqrt(mean_squared_error(Y_test, pred)))
print("R2:", r2_score(Y_test, pred))
```

HYPER PARAMETER TUNING


```python
from sklearn.model_selection import GridSearchCV
```
