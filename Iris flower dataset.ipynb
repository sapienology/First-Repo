```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("/content/IRIS.csv")
```

```python
 df1
```

```python
df1.info()
```

```python
df1['species'].unique()
```

```python
df1['species'] = df1['species'].replace({
    'Iris-setosa' : 0 ,
    'Iris-versicolor' : 1 ,
    'Iris-virginica' : 2
})
```

```python
df1
```

TEST TRAIN SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
x = df1.drop(columns = ['species'])
```

```python
y = df1['species']
```

```python
xtrain ,xtest , ytrain, ytest = train_test_split(x,y , test_size = 0.2 , random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lu = LogisticRegression()
```

```python
lu.fit(xtrain , ytrain)
```

```python
pre = lu.predict(xtest)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,pre))
```

HYPERPARAMETER TUNING

```python
from sklearn.model_selection import GridSearchCV
```

```python
log_param = {'C' : [0.1 ,1] ,  'penalty' : ['11' , 'l2']}
```

```python
grid = GridSearchCV(
    LogisticRegression(),
    log_param
)
```

```python
grid.fit(xtrain , ytrain)
```

```python
GridSearchCV(estimator=LogisticRegression(),
             param_grid={'C': [0.1, 1], 'penalty': ['l1', 'l2']})
```

```python
grid.best_estimator_
```

```python

LogisticRegression(C=0.1)
```

```python


grid_pred = grid.predict(xtest)
```

```python
print(classification_report(ytest,grid_pred))
```

SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
smut = SMOTE(random_state = 42)
```

```python
ytrain.value_counts()
```

```python
xsample , ysample = smut.fit_resample(x,y)
```

```python
ysample.value_counts()
```

```python
xtrain_new , xtest_new ,ytrain_new , ytest_new = train_test_split(xsample ,ysample , test_size = 0.2 , random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
```

```python
scaled_x_train = scale.fit_transform(xtrain_new)
```

```python
scaled_x_test = scale.transform(xtest_new)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(scaled_x_train , ytrain_new)
```

```python
pred = log.predict(scaled_x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest_new,pred))
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 3  , init = 'k-means++' ,
               n_init =10 , max_iter = 300 ,
               random_state = 42)
```

```python
model.fit(scaled_x_train)
```

```python
pe = model.predict(scaled_x_test)
```

```python
print(classification_report(ytest_new,pe))
```
