```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('/content/amazon_products_sales_data_uncleaned.csv')
```

Bought in last month Column

```python
df1['bought_in_last_month'].unique()
```

```python
lowered = df1['bought_in_last_month'].astype(str).str.lower()
lowered = lowered.str.replace(',', '', regex=False)

numbers = pd.to_numeric(lowered.str.extract(r'(\d+\.?\d*)')[0],
                        errors = 'coerce')
multiplier = np.where(lowered.str.contains('k') , 1000 ,1)

df1['bought_in_last_month'] = (numbers * multiplier).round().astype('Int64')
```

```python
df1 = df1.dropna(subset = ['bought_in_last_month'])
```

```python
df1.isna().sum()
```

Rating COlumn

```python
df1['rating'] = df1['rating'].astype(str).str.extract(r'(\d+\.?\d*)')[0]
```

```python
df1['rating'] = df1['rating'].astype(float)
```

```python
df1['rating'] = df1['rating'].fillna(df1['rating'].mean())
```
